# Vision Model Comparison

Compare synthetic baseline behavior across standard vision backbones and a CLIP image encoder.

```python
# This setup cell has no meaningful table output.
# It imports the runner and data tools used by the later cells.

from pathlib import Path
import json
import pandas as pd

from vision.model_comparison import DEFAULT_MODEL_NAMES
from vision.run_model_comparison import run_model_comparison
```

```python
# This cell prints the model list that will be compared.
# The output is just the benchmark configuration: which backbones, device, image size,
# and random seed are being used for the synthetic comparison.

MODEL_NAMES = list(DEFAULT_MODEL_NAMES)
DEVICE = "cpu"
BATCH_SIZE = 16
IMAGE_SIZE = 224
SEED = 0
OUTPUT_DIR = Path("results/vision/model_comparison")

MODEL_NAMES
```

```python
# This cell runs the full synthetic model-comparison benchmark.
# The returned summary tells you whether all sanity checks passed, which rows gave the best
# identification and generalization, and whether any tradeoff candidates were found.
# Main finding in this notebook: no model produced the desired tradeoff curve.

summary = run_model_comparison(
    model_names=MODEL_NAMES,
    device=DEVICE,
    batch_size=BATCH_SIZE,
    image_size=IMAGE_SIZE,
    seed=SEED,
    output_dir=OUTPUT_DIR,
)
summary
```

```python
# This table shows raw rows from the combined synthetic benchmark CSV.
# Each row is one experimental condition: model, task_mode, pooling, layer, storage size,
# and the resulting identification/generalization metrics.
# Useful fields:
# - ident_accuracy / gen_accuracy: retrieval accuracy for the two tasks
# - ident_avg_margin / gen_avg_margin: how far the correct target is from the best wrong item
# - avg_margin: average of the identification and generalization margins
# Read this as the row-level data behind the later summary tables.

combined = pd.read_csv(OUTPUT_DIR / "model_comparison_combined.csv")
combined.head()
```

```python
# These two tables show the best row per model under two criteria.
# best_ident: for each model, the condition with the highest identification accuracy.
# best_gen: for each model, the condition with the highest generalization accuracy.
# Fields:
# - task_mode: synthetic task family (mixed_color_position, color_only, position_only)
# - pooling: feature readout used for the backbone
# - layer: representation layer being evaluated
# - n_stored_per_color: how many stored examples per color were used
# - ident_accuracy / gen_accuracy: percent correct on the two tasks
# - avg_margin: average separation of the correct target from competitors
# tradeoff_candidates is the key result: an empty list means the desired synthetic tradeoff
# did not appear in this model-comparison pass.

summary_path = OUTPUT_DIR / "model_comparison_summary.json"
summary_json = json.loads(summary_path.read_text())

best_ident = pd.DataFrame(summary_json["best_identification"]).T[
    ["task_mode", "pooling", "layer", "n_stored_per_color", "ident_accuracy", "gen_accuracy", "avg_margin"]
]
best_gen = pd.DataFrame(summary_json["best_generalization"]).T[
    ["task_mode", "pooling", "layer", "n_stored_per_color", "ident_accuracy", "gen_accuracy", "avg_margin"]
]

display(best_ident)
display(best_gen)
summary_json["tradeoff_candidates"]
```

```python
# This table directly compares ViT and CLIP on the synthetic benchmark.
# Rows are grouped by model, task_mode, pooling, and layer, then averaged over the other
# synthetic settings such as storage size and perturbation condition.
# Fields:
# - ident_accuracy / gen_accuracy: mean performance for that grouped condition
# - ident_avg_margin / gen_avg_margin: mean margin on each task
# - avg_margin: overall mean margin across both tasks
# Main finding: CLIP behaves similarly on easy color-only conditions but does not fix the
# harder synthetic mixed or position-heavy settings.

vit_vs_clip = combined[combined["model_name"].isin(["vit_base_patch16_224", "vit_base_patch16_clip_224.openai"])]
vit_vs_clip.groupby(["model_name", "task_mode", "pooling", "layer"]).agg({
    "ident_accuracy": "mean",
    "gen_accuracy": "mean",
    "ident_avg_margin": "mean",
    "gen_avg_margin": "mean",
    "avg_margin": "mean",
}).sort_values(["task_mode", "gen_accuracy"], ascending=[True, False]).head(18)
```
